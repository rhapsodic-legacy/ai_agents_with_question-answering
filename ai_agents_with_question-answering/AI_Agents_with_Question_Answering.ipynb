```python
# Enhanced AI Agents with Question-Answering

# Optimized for Google Colab (Free Tier)

# --- Introduction ---
# This notebook enhances an AI agent using LangChain for factual question-answering, refines responses.
# for complex questions, tags subjective questions, and includes a Streamlit UI with ngrok for Colab compatibility.

# --- Step 1: Install Dependencies ---
!pip install --upgrade torch torchvision -q
!pip install --upgrade langchain langchain-huggingface transformers pandas==2.2.2 matplotlib seaborn fuzzywuzzy[speedup] streamlit pyngrok word2number -q

# --- Step 2: Import Libraries ---
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, pipeline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
from fuzzywuzzy import fuzz
from word2number import w2n
import os
import logging
import warnings
import streamlit as st
from pyngrok import ngrok
import json

# Suppress warnings.
os.environ["TRANSFORMERS_VERBOSITY"] = "error"
logging.getLogger("transformers").setLevel(logging.ERROR)
warnings.filterwarnings("ignore", category=FutureWarning)

# --- Step 3: Set Up Local QA Pipeline ---
print("Setting up local QA pipeline...")
model_name = "deepset/roberta-base-squad2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
hf_pipeline = pipeline(
    "question-answering",
    model=model,
    tokenizer=tokenizer,
    device=0 if torch.cuda.is_available() else -1
)
llm = HuggingFacePipeline(pipeline=hf_pipeline)

# --- Step 4: Define Dataset ---
print("Defining dataset with edge cases...")

data = pd.DataFrame(robust_qa_dataset)

# --- Step 5: Tag Subjective Questions ---
print("Tagging subjective questions...")
def is_subjective(true_answer):
    subjective_phrases = ["subjective", "no definitive answer", "depends", "varies", "personal"]
    return any(phrase in str(true_answer).lower() for phrase in subjective_phrases)

data["is_subjective"] = data["true_answer"].apply(is_subjective)

# --- Step 6: Annotate LLM Outputs ---
print("Annotating LLM outputs...")
def get_answer_from_pipeline(row):
    result = hf_pipeline({"question": row["question"], "context": row["context"]})
    answer = result['answer'].lower().strip() if result['answer'] else ""
    score = result['score']
    return answer, score

data[["initial_response", "initial_pipeline_score"]] = data.apply(
    lambda row: pd.Series(get_answer_from_pipeline(row)), axis=1
)

def normalize_answer(answer):
    """Convert word-based numbers to digits and normalize text."""
    if isinstance(answer, str):
        try:
            # Convert word numbers to digits (e.g., "seven" -> "7").
            answer = str(w2n.word_to_num(answer))
        except ValueError:
            # If not a number word, just lowercase and strip.
            answer = answer.lower().strip()
    return answer

def assign_score(response, true_answer, is_subjective):
    if is_subjective:
        return 1.0 if response else 0.5
    if not isinstance(response, str) or not response:
        return 0.5
    response = normalize_answer(response)
    true_answer = normalize_answer(true_answer)
    return 0.9 if fuzz.partial_ratio(response, true_answer) >= 70 else 0.7

data["initial_score"] = data.apply(
    lambda row: assign_score(row["initial_response"], row["true_answer"], row["is_subjective"]), axis=1
)
data["initial_correct"] = data.apply(
    lambda row: row["is_subjective"] or fuzz.partial_ratio(
        normalize_answer(row["initial_response"]), normalize_answer(row["true_answer"])
    ) >= 70, axis=1
)

# --- Step 7: Refine Responses ---
print("Refining responses for complex or low-confidence questions...")
def refine_response(row):
    complex_keywords = ["three", "list", "main", "parts", "reasons", "all"]
    if row["initial_pipeline_score"] < 0.5 or any(keyword in row["question"].lower() for keyword in complex_keywords):
        # Enhanced prompt for list-based questions.
        refined_question = f"Provide the full list or complete answer to: {row['question']}"
        result = hf_pipeline({"question": refined_question, "context": row["context"]})
        refined_resp = result['answer'].lower().strip() if result['answer'] else ""
        return refined_resp, assign_score(refined_resp, row["true_answer"], row["is_subjective"])
    return row["initial_response"], row["initial_score"]

data[["refined_response", "refined_score"]] = data.apply(refine_response, axis=1, result_type="expand")
data["refined_correct"] = data.apply(
    lambda row: row["is_subjective"] or fuzz.partial_ratio(
        normalize_answer(row["refined_response"]), normalize_answer(row["true_answer"])
    ) >= 70, axis=1
)

# --- Step 8: Performance Report ---
print("Generating performance report...")
objective_data = data[~data["is_subjective"]]
initial_accuracy = objective_data["initial_correct"].mean()
refined_accuracy = objective_data["refined_correct"].mean()
print(f"Initial Accuracy (Objective Questions): {initial_accuracy:.2f}")
print(f"Refined Accuracy (Objective Questions): {refined_accuracy:.2f}")

# --- Calculate Additional Metrics ---
print("Calculating additional metrics (Precision, Recall, F1)...")
def calculate_metrics(true_answer, model_response, threshold=70):
    true_items = [item.strip().lower() for item in true_answer.replace("and", ",").split(",")]
    model_items = [item.strip().lower() for item in model_response.replace("and", ",").split(",")]

    tp = sum(1 for true_item in true_items if any(fuzz.partial_ratio(true_item, model_item) >= threshold for model_item in model_items))
    fn = len(true_items) - tp
    fp = sum(1 for model_item in model_items if not any(fuzz.partial_ratio(model_item, true_item) >= threshold for true_item in true_items))

    return tp, fp, fn

# Initialize totals.
total_tp_initial = total_fp_initial = total_fn_initial = 0
total_tp_refined = total_fp_refined = total_fn_refined = 0

# Compute metrics for each objective question.
for index, row in objective_data.iterrows():
    tp_i, fp_i, fn_i = calculate_metrics(row["true_answer"], row["initial_response"])
    total_tp_initial += tp_i
    total_fp_initial += fp_i
    total_fn_initial += fn_i

    tp_r, fp_r, fn_r = calculate_metrics(row["true_answer"], row["refined_response"])
    total_tp_refined += tp_r
    total_fp_refined += fp_r
    total_fn_refined += fn_r

# Calculate precision, recall, F1.
precision_initial = total_tp_initial / (total_tp_initial + total_fp_initial) if total_tp_initial + total_fp_initial > 0 else 0
recall_initial = total_tp_initial / (total_tp_initial + total_fn_initial) if total_tp_initial + total_fn_initial > 0 else 0
f1_initial = 2 * (precision_initial * recall_initial) / (precision_initial + recall_initial) if precision_initial + recall_initial > 0 else 0

precision_refined = total_tp_refined / (total_tp_refined + total_fp_refined) if total_tp_refined + total_fp_refined > 0 else 0
recall_refined = total_tp_refined / (total_tp_refined + total_fn_refined) if total_tp_refined + total_fn_refined > 0 else 0
f1_refined = 2 * (precision_refined * recall_refined) / (precision_refined + recall_refined) if precision_refined + recall_refined > 0 else 0

# Print additional metrics.
print("\nAdditional Metrics for Objective Questions:")
print(f"Initial Precision: {precision_initial:.2f}")
print(f"Initial Recall: {recall_initial:.2f}")
print(f"Initial F1 Score: {f1_initial:.2f}")
print(f"Refined Precision: {precision_refined:.2f}")
print(f"Refined Recall: {recall_refined:.2f}")
print(f"Refined F1 Score: {f1_refined:.2f}")

# Save metrics to JSON.
metrics = {
    "initial_accuracy": initial_accuracy,
    "refined_accuracy": refined_accuracy,
    "precision_initial": precision_initial,
    "recall_initial": recall_initial,
    "f1_initial": f1_initial,
    "precision_refined": precision_refined,
    "recall_refined": recall_refined,
    "f1_refined": f1_refined
}
with open("metrics.json", "w") as f:
    json.dump(metrics, f)

data.to_csv("annotated_qa_outputs.csv", index=False)
print("Saved annotated data to 'annotated_qa_outputs.csv'")

# --- Step 9: Visualize Results ---
print("Visualizing response quality...")
fig, ax = plt.subplots(figsize=(10, 6))
metrics_list = ["Accuracy", "Precision", "Recall", "F1 Score"]
initial_values = [initial_accuracy, precision_initial, recall_initial, f1_initial]
refined_values = [refined_accuracy, precision_refined, recall_refined, f1_refined]
x = range(len(metrics_list))
ax.bar([i - 0.2 for i in x], initial_values, width=0.4, label="Initial", color="blue")
ax.bar([i + 0.2 for i in x], refined_values, width=0.4, label="Refined", color="orange")
ax.set_xticks(x)
ax.set_xticklabels(metrics_list)
ax.set_ylabel("Score")
ax.set_title("Performance Metrics: Initial vs. Refined")
ax.legend()
plt.savefig("performance_metrics.png")
plt.show()

# --- Step 10: Streamlit UI Integration with ngrok ---
print("Setting up Streamlit UI with ngrok...")

# Install ngrok.
!wget https://bin.equinox.io/c/bNyj1mQVY4c/ngrok-v3-stable-linux-amd64.tgz
!tar -xvf ngrok-v3-stable-linux-amd64.tgz
!mv ngrok /usr/local/bin

# Get your ngrok auth token from https://dashboard.ngrok.com/get-started/your-authtoken
NGROK_AUTH_TOKEN = "YOUR_KEY_HERE"  # Replace with your token.
!ngrok authtoken $NGROK_AUTH_TOKEN

# Streamlit app code.
streamlit_code = """
import streamlit as st
from transformers import pipeline
import json

# Load the QA pipeline.
hf_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

st.title("AI Question-Answering Agent")
st.write("Enter a question and context to get an answer:")

question = st.text_input("Question")
context = st.text_area("Context")

if st.button("Get Answer"):
    if question and context:
        initial_result = hf_pipeline({"question": question, "context": context})
        initial_answer = initial_result['answer'].lower().strip() if initial_result['answer'] else ""
        initial_score = initial_result['score']

        refined_answer = initial_answer
        if initial_score < 0.5 or any(keyword in question.lower() for keyword in ["three", "list", "main", "parts", "reasons", "all"]):
            refined_question = f"Provide the full list or complete answer to: {question}"
            refined_result = hf_pipeline({"question": refined_question, "context": context})
            refined_answer = refined_result['answer'].lower().strip() if refined_result['answer'] else ""

        st.write(f"**Initial Answer:** {initial_answer} (Confidence: {initial_score:.2f})")
        st.write(f"**Refined Answer:** {refined_answer}")
    else:
        st.write("Please enter both a question and context.")

# Load and display metrics.
try:
    with open('metrics.json', 'r') as f:
        metrics = json.load(f)
    st.sidebar.header("Overall Performance Metrics")
    st.sidebar.write(f"Initial Accuracy: {metrics['initial_accuracy']:.2f}")
    st.sidebar.write(f"Refined Accuracy: {metrics['refined_accuracy']:.2f}")
    st.sidebar.write(f"Initial Precision: {metrics['precision_initial']:.2f}")
    st.sidebar.write(f"Initial Recall: {metrics['recall_initial']:.2f}")
    st.sidebar.write(f"Initial F1 Score: {metrics['f1_initial']:.2f}")
    st.sidebar.write(f"Refined Precision: {metrics['precision_refined']:.2f}")
    st.sidebar.write(f"Refined Recall: {metrics['recall_refined']:.2f}")
    st.sidebar.write(f"Refined F1 Score: {metrics['f1_refined']:.2f}")
except FileNotFoundError:
    st.sidebar.write("Metrics not available.")
"""

# Save Streamlit app to a file.
with open("app.py", "w") as f:
    f.write(streamlit_code)

# Run Streamlit app and set up ngrok tunnel.
!streamlit run app.py &>/dev/null &
public_url = ngrok.connect(8501)
print(f"Streamlit app is live at: {public_url}")

# --- Step 11: Summary ---
print("\nSummary:")
print(f"- Dataset ready for your 100 questions (add tomorrow) with 50 edge cases included.")
print(f"- Initial accuracy: {initial_accuracy:.2f}, refined accuracy: {refined_accuracy:.2f}.")
print(f"- Additional metrics - Initial: Precision={precision_initial:.2f}, Recall={recall_initial:.2f}, F1={f1_initial:.2f}; Refined: Precision={precision_refined:.2f}, Recall={recall_refined:.2f}, F1={f1_refined:.2f}.")
print("- UI integrated with Streamlit and ngrok for interactive use.")
```
